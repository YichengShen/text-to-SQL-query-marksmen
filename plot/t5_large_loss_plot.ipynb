```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
train_loss_csv = "/content/drive/Shareddrives/CIS 522/t5-large-loss-data/training_loss.csv"
val_loss_csv = "/content/drive/Shareddrives/CIS 522/t5-large-loss-data/val_loss.csv"

train_loss_df = pd.read_csv(train_loss_csv)
val_loss_df = pd.read_csv(val_loss_csv)
```

```python
df = train_loss_df[['Value']]
df = df.rename(columns={'Value': 'training loss'})
df['validation loss'] = val_loss_df['Value']
df['epoch'] = range(1,9)
```

```python
df.plot(x ='epoch', color=['deeppink', 'purple'])

plt.axvline(x=4, color='r', linestyle='dotted', label='Epoch 4')

plt.xlabel('Epoch')
plt.ylabel('CrossEntropy loss')
plt.savefig('loss_plot.png', dpi=800)
plt.show()
```
